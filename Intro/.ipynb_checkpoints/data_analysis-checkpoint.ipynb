## Dimensionality Reduction and Feature Extraction

Some useful links:

```python

# Create useful plotting functions

def plot_components(X_reduced, y, comp1=0, comp2=1):
    # Plot component of decompositions
    
    labels_dict = {label:i for i, label in enumerate(y.unique())}
    labels = [labels_dict[label] for label in y]
    
    exp_var = np.var(X_reduced, axis=0)/np.sum(np.var(X_reduced, axis=0))
    print(f'Explained Variance Ratio: {exp_var}')

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.scatter(X_reduced[:, comp1], X_reduced[:, comp2], c=labels, cmap=plt.cm.Set1, edgecolor='k')
    ax.set_title(f"{algorithm_name} components", fontsize=16)
    ax.set_xlabel(f"Eigen-vector {comp1+1}", fontsize=16)
    ax.set_ylabel(f"Eigen-vector {comp2+1}", fontsize=16)
    plt.show()

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram
    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
    plt.xlabel("Number of points in node (or index of point if no parenthesis).")
    plt.show()
        
def plot_clusters(data, labels, algorithm_name):
    palette = sns.color_palette('deep', np.unique(labels).max() + 1)
    colors = [palette[x] if x >= 0 else (0.0, 0.0, 0.0) for x in labels]
    
    plt.scatter(data.T[0], data.T[1], c=colors)
    frame = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)
    plt.title(f'Clusters found by {algorithm_name}', fontsize=18)
    plt.show()

############################################################

n_components = 3
n_clusters = 2

clustering_algorithm = {"PCA": PCA(n_components=n_components),
                        "NMF": NMF(n_components=n_components),
                        "SparsePCA": SparsePCA(n_components=n_components),
                        "FastICA": FastICA(n_components=n_components), 
                        "FACTOR": FactorAnalysis(n_components=n_components),
                        "KMeans": KMeans(n_clusters=n_clusters),
                        "TSNE": TSNE(n_components=3, random_state=42),
                        "ISOMAP": Isomap(n_components=n_components),
                        "UMAP": UMAP(n_components=n_components, random_state=42),
                        "AggloClust": AgglomerativeClustering(distance_threshold=None, n_clusters=2),
                        "DBSCAN": DBSCAN(eps=0.03, min_samples=4),
                        "HDBSCAN": HDBSCAN(min_samples=3, min_cluster_size=20),                
                        }
```
