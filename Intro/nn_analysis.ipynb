# Neural Network Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.utils import resample

from keras.models import Sequential, Model
from keras.layers import Dense, Dropout, Lambda, Flatten, Activation, Reshape
from keras.optimizers import Adam, RMSprop, SGD
from keras.preprocessing.image import ImageDataGenerator
from keras.utils import np_utils
from keras import  backend as K

# CNN
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization, UpSampling2D

from keras.datasets import mnist

#from keras.layers.core import Dense

import theano 
```

Example from [here](https://www.bogotobogo.com/python/scikit-learn/Artificial-Neural-Network-ANN-10-Deep-Learning-3-Theano-TensorFlow-Keras.php).

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train.shape, y_train.shape
```

```python
# image in 64-bit format

theano.config.floatX = 'float64'
X_train = X_train.astype(theano.config.floatX)
X_test = X_test.astype(theano.config.floatX)
X_train.shape, X_test.shape

fig, ax = plt.subplots(1, 3, figsize=(8, 4))
for i in range(0, 3):
    ax[i].imshow(X_train[i], cmap=plt.get_cmap('gray'))
    ax[i].set_title(y_train[i]);
```

**Reshape**: We're going to use a Dense Neural Network Architecture, which takes 1D data structures, not 2D images. So we need to make the input shape appropriate, so we 'flatten' it.

**One Hot Encoding**: This increases the efficiency of the matrix algebra during network training and evaluation.

```python
# flatten images
img_rows, img_cols = X_train[0].shape[0], X_train[0].shape[1] 
img_size = img_rows*img_cols
X_train = X_train.reshape(X_train.shape[0], img_size)
X_test = X_test.reshape(X_test.shape[0], img_size)
print(X_train.shape)

# scale data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

# one-hot encoding class labels

y_test_OneHotEncoding = np_utils.to_categorical(y_test) 
y_test_OneHotEncoding[:10]
y_train_OneHotEncoding = np_utils.to_categorical(y_train) 
y_train_OneHotEncoding[:10]
```

```python
np.random.seed(42) 

model = Sequential()  # to implement a feedforward neural network
model.add(Dense(50,  # output dimension
                input_dim=X_train.shape[1], 
                kernel_initializer='uniform', 
                activation='relu'))  # relu, tanh or sigmoid

model.add(Dense(50,
                input_dim=50, 
                kernel_initializer='uniform', 
                activation='relu'))

model.add(Dense(y_train_OneHotEncoding.shape[1],
                input_dim=50, 
                kernel_initializer='uniform', 
                activation='softmax'))

model.summary()
```

## **Compile the model**

Select three key options

1. **optimizer**: the method for optimizing the weights. "Stochastic Gradient Descent (SGD)" is the canonical method.
2. **loss function**: the form of the function to encode the difference between the data's true label and the predict label.
3. **metric**: the function by which the model is evaluated.

```python
sgd = SGD(lr=0.001, decay=1e-7, momentum=.9)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])

history = model.fit(X_train, y_train_OneHotEncoding, epochs=20, 
                      batch_size=300, verbose=1, 
                      validation_split=0.2,
                      workers=4, use_multiprocessing=True)
```

## **Diagnostics!**

Because neural networks are such complex objects, and in this era of experimentation, diagnostics (both standard and tailored) are critical for training NN models.

### Evaluate overall model efficacy

Evaluate model on training and test data and compare. This provides summary values that are equivalent to the final value in the accuracy/loss history plots.


```python
loss_train, acc_train  = model.evaluate(X_train, y_train_OneHotEncoding, verbose=False)
loss_test, acc_test  = model.evaluate(X_test, y_test_OneHotEncoding, verbose=False)
print(f'Train acc/loss: {acc_train:.3}, {loss_train:.3}')
print(f'Test acc/loss: {acc_test:.3}, {loss_test:.3}')
```

```python
y_pred_train = model.predict(X_train, verbose=True)
y_pred_test = model.predict(X_test,verbose=True)
```

```python
# set up figure
f = plt.figure(figsize=(12,5))
f.add_subplot(1,2, 1)

# plot accuracy as a function of epoch
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')

# plot loss as a function of epoch
f.add_subplot(1,2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show(block=True)
```

## **Confusion Matrix**

```python
labels = np.unique(y_test_OneHotEncoding.argmax(axis=1))
matrix = confusion_matrix(y_test_OneHotEncoding.argmax(axis=1), y_pred_test.argmax(axis=1), labels, normalize='true')
```

```python
figure = plt.figure(figsize=(8, 8))
sns.heatmap(matrix, annot=True, cmap=plt.cm.Blues)
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
autoencoder = Sequential()

# Encoder Layers
autoencoder.add(Conv2D(16, (3, 3), activation='relu', padding='same', input_shape=X_train.shape[1:]))
autoencoder.add(MaxPooling2D((2, 2), padding='same'))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(MaxPooling2D((2, 2), padding='same'))
autoencoder.add(Conv2D(8, (3, 3), strides=(2,2), activation='relu', padding='same'))
autoencoder.add(MaxPooling2D((2, 2), padding='same'))
autoencoder.add(Conv2D(8, (3, 3), strides=(2,2), activation='relu', padding='same'))
autoencoder.add(MaxPooling2D((2, 2), padding='same'))

# Flatten encoding for visualization
autoencoder.add(Flatten())
autoencoder.add(Reshape((1, 1, 8)))

# Decoder Layers
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(16, (3, 3), activation='relu'))
autoencoder.add(UpSampling2D((2, 2)))
autoencoder.add(Conv2D(1, (3, 3), activation='sigmoid', padding='same'))

autoencoder.summary()
```

```python
# for visualisation
encoder = Model(inputs=autoencoder.input, outputs=autoencoder.get_layer('flatten_5').output)
encoder.summary()
```
